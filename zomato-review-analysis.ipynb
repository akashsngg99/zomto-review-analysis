Making the reviews single line

```python
review=[]
for i in range(0, 100):
  r=[]
  r1=mydataset['reviews_list'][i]
  r=r1.split(")")
  del r[-1]
  review.extend(r)
```

```python
review
```

Removing special chars and converting to meaningful lines

```python
import re

for i in range(len(review)):
  s=review[i]
  m=re.search("RATED",s)
  if(m):
    s=s[(m.end()+4):]
  
  s=re.sub("(n\'t)|(\w+nt)"," not",s)
  s=re.sub("(\'ve)"," have",s)
  s=re.sub("(\'s)"," is",s)
  s=re.sub("(\'m)"," am",s)
  s=re.sub("\\\\n","  ",s)
  s=re.sub(".\\\\.","  ",s)
  s=re.sub("\\\\.","  ",s)
  s=re.sub("\\\\x","  ",s)
  s=re.sub("x8","  ",s)
  s=re.sub('[^a-zA-Z]',' ',s)

  review[i]=s
```

```python
import re
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
c=[]
for i in range(0,2694):
    review = re.sub('[^a-zA-Z]', ' ', sampleData['reviews'][i])
    review=review.lower()
    review=review.split()
    review = [word for word in review if not word in set(stopwords.words('english'))]
    
    #applystemming concept
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    # we use same line as there may be word left which are stopword
    review=' '.join(review)
    c.append(review)
```

```python
c
```

Making reviview and label file csv

```python
labels=[]
for i in range(0,2694):
    review = re.sub('[^a-zA-Z]', ' ', sampleData['label'][i])
    labels.append(review)
```

```python
len(labels)
```

saving the reviews and labels

```python
import csv
with open('zomato50000_reviews.csv', 'w') as myfile:
  wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
  wr.writerow(c)
```

```python
import csv
with open('zomato50000_labels.csv', 'w') as myfile:
  wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
  wr.writerow(labels)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=1500)
x=cv.fit_transform(c).toarray()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()

y=lb.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(input_dim=1500, init="random_uniform", activation='sigmoid',output_dim=1000))
model.add(Dense(output_dim=1000, init="random_uniform", activation='sigmoid'))
model.add(Dense(output_dim=3, init="random_uniform", activation='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train,y_train, epochs=10, batch_size=10)
```

```python
model.save("newmyNLPmodel.h5")
```
